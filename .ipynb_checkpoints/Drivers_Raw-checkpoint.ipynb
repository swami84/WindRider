```python
drivers = pd.read_csv('C:/Machine Learning/TNP_Drivers.csv',low_memory=False)

tic = process_time()

drivers = drivers.dropna(subset=['DRIVER_START_MONTH','ZIP'])

drivers=drivers[drivers['NUMBER_OF_TRIPS']!=0]
drivers['NUMBER_OF_TRIPS'] = drivers['NUMBER_OF_TRIPS'].fillna(1500)


drivers = drivers.reset_index(drop=True)
drivers['ZIP'] = drivers['ZIP'].apply(lambda x: pd.to_numeric(x, errors='coerce')).fillna(0)
drivers['ZIP']=drivers['ZIP'].astype(int)
drivers = drivers.loc[ (drivers['ZIP']>59999)&(drivers['ZIP']<=69999)]
drivers['Rep_Month']= pd.to_datetime(drivers['MONTH_REPORTED'], format='%Y-%m')
drivers['Start_Month']= pd.to_datetime(drivers['DRIVER_START_MONTH'], format='%Y-%m')
drivers = drivers.drop(columns=['MONTH_REPORTED','DRIVER_START_MONTH'])
drivers['Days_Driving'] = pd.to_datetime(drivers['Rep_Month']).dt.date - pd.to_datetime(drivers['Start_Month']).dt.date
drivers['Months_Driving'] = ((drivers['Days_Driving'].dt.days)/30).astype(int)
drivers['Days_Driving'] = drivers['Days_Driving'].dt.days
drivers['Start_Year']= drivers['Start_Month'].dt.year
drivers['Rep_Year']= drivers['Rep_Month'].dt.year
drivers['Years_Driving'] = drivers['Rep_Year']-drivers['Start_Year']

#drivers = drivers[drivers['Years_Driving']>=0]
#drivers = drivers[drivers['Months_Driving']>=0]
drivers= drivers.reset_index(drop=True)
toc = process_time()
print('Process Time for Data Cleaning = ', (toc-tic)/60, 'mins')
drivers.info()
```

```python

lat =[]
lng = []
for i in range(0,len(drivers)):
    zip_info = search.by_zipcode(drivers['ZIP'][i])
    lat.append(json.loads(zip_info.to_json())['lat'])
    lng.append(json.loads(zip_info.to_json())['lng'])
    if ((i%250000)==0):
        print ('Processed', i, 'rows')
drivers['Latitude'] = lat
drivers['Longitude'] = lng
drivers = drivers.dropna(subset=['Latitude', 'Longitude'])


drivers = drivers.reset_index()

def haversine_np(lon1, lat1, lon2, lat2):
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2

    c = 2 * np.arcsin(np.sqrt(a))
    distance = 3958.8 * c
    return distance

drivers['Distance'] = haversine_np( -87.6298,41.8781,drivers['Longitude'],drivers['Latitude'])
drivers.to_csv('drivers.csv',index=False, sep='\t')
#drivers=drivers.loc[drivers['Distance']<100]
```

```python
drivers_prep = pd.read_csv('drivers.csv', low_memory=False,sep='\t')
drivers_prep=drivers_prep.loc[drivers_prep['Distance']<25]
drivers_prep =drivers_prep.loc[drivers_prep['Start_Year']>2012]
drivers_prep['Months_Driving'] = drivers_prep['Months_Driving'] + 1
drivers_prep['Trips_per_Month']  = drivers_prep['NUMBER_OF_TRIPS'] /drivers_prep['Months_Driving'] 
drivers_prep["Years_Driving"] = drivers_prep["Months_Driving"]/12
#drivers_prep=drivers_prep.loc[drivers_prep['Years_Driving']<=1]
drivers_prep['Trips_per_Year']  = drivers_prep['NUMBER_OF_TRIPS'] /drivers_prep['Years_Driving']
```

```python
zip_summary=drivers_prep.groupby(['ZIP','Start_Year'])['NUMBER_OF_TRIPS'].agg(['count']).sort_values(['count'], ascending=False)
zip_summary['Sum_Trips'] = drivers_prep.groupby(['ZIP','Start_Year'])['NUMBER_OF_TRIPS'].agg(['sum']).sort_values(['sum'])
zip_summary['Avg_Months_Driving'] = drivers_prep.groupby(['ZIP','Start_Year'])['Months_Driving'].agg(['mean']).sort_values(['mean'])
zip_summary['Avg_Trips_PM'] = zip_summary['Sum_Trips']/(zip_summary['count']*zip_summary['Avg_Months_Driving'])
zip_summary.sort_values(by=['Avg_Trips_PM'], ascending=False)
```
